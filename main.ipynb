```python
from langchain_community.document_loaders import UnstructuredPowerPointLoader


powerpoint_loader = UnstructuredPowerPointLoader("data/pptx/DDBS Lecture1.ppt")

data = powerpoint_loader.load()

# data
```

```python
# print(data[0].page_content)
pages = data[0].page_content.split("<number>")
cleaned_pages = [page.strip() for page in pages]
# Output the result to check the separated pages

print("Filename: ", data[0].metadata["source"])
print("-----------------------")
for i, page in enumerate(cleaned_pages):
    print(f"Page {i+1}:")
    print(page)
    print("\n" + "=" * 50 + "\n")
```
